```python
from KRData.HKData import HKFuture
import pandas as pd
import datetime as dt
import talib
hf = HKFuture('krdata', 'kairuitouzi','www.a667.com')
```

```python
trade_dates = hf.get_main_contract_trade_dates('HSI', '20150101', '20190401')    #选择合同日期
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
trade_results = []
close_results = []
for code, tds in trade_dates.items():
#     print(f'Backtest{code}- From:{tds[0]} To:{tds[-1]}')
    s = tds[0]
    e = tds[-1]
    df = hf.get_bars(code, fields=[ 'open', 'high', 'low', 'close', 'trade_date'], start=s, end=e)
    df = df.drop_duplicates(subset=["trade_date"],keep="last")
    trade_results = trade_results + list(df['trade_date'].values)
    close_results = close_results + list(df['close'].values)
```

```python
all = {'close':close_results, 'trade_results':trade_results}
table = pd.DataFrame(all)
table['MA60'] = talib.MA(table['close'], timeperiod=60)      #显示60均线
table
```

```python
def jian(close, ma60):      #突破60线买，低于60线做空
    p_and_l = []
    b_s = []
    cs= 0
    for i in range(len(close)):
        
        if str(ma60[i]) == 'nan':
            p_and_l.append(ma60[i])
            b_s.append(ma60[i])
        else:
            num = '%.2f'%((close[i]-ma60[i])/ma60[i]*100)
            num = float(num)
            p_and_l.append((str(num)+'%'))
            if num < 0:
                b_s.append('Sell')
            elif num >= 0:
                b_s.append('Buy')

            cs+=1
    return p_and_l , b_s
p_and_l , b_s = jian(table['close'], table['MA60'])
```

```python
table['Profit and loss'] = p_and_l
table['Buy_Sell'] = b_s
table[['trade_results','close','MA60','Profit and loss', 'Buy_Sell']]
```
