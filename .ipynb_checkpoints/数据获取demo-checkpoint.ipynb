```python
from KRData.IBData import IBMarket
import datetime as dt
import jqdatasdk
import talib
import pandas as pd
jqdatasdk.auth('13580726532', '127565568yjd')
```

```python
ibm = IBMarket()
```

```python
def getData():
    ibdata = ibm.get_bars_from_ib('HSI1907', dt.datetime.now().replace(hour=0, minute=31, second=0), 
                                  dt.datetime.now().replace(hour=15, minute=0, second=0))
    jqdata = jqdatasdk.get_price('IF9999.CCFX', start_date=dt.datetime.now().replace(hour=9, minute=31, second=0)- dt.timedelta(days=1), 
                                 end_date=dt.datetime.now().replace(hour=15, minute=0, second=0), frequency='1m', fields=None, skip_paused=False, fq=None)
    jqdata.index = jqdata.index - dt.timedelta(minutes=1)


    ibdata['ma60'] = talib.MA(ibdata['close'].values, timeperiod=60)
    ibdata['bias_HSI'] = (ibdata['close'].values - ibdata['ma60'].values)/ibdata['ma60'].values*100.0

    jqdata['ma60'] = talib.MA(jqdata['close'].values, timeperiod=60)
    jqdata['bias_IF'] = (jqdata['close'].values - jqdata['ma60'].values)/jqdata['ma60'].values*100.0
    return ibdata,jqdata
```

```python
cont2='2019/7/8 14:12'
ibdata,jqdata=getData()
a=ibdata[ibdata.index>cont2]['bias_HSI'].agg(['sum','mean','count'])
b=jqdata[(jqdata.volume>0) & (jqdata.index>cont2)]['bias_IF'].agg(['sum','mean','count'])
pd.concat([a,b],axis=1)
```

```python
ibdata,jqdata=getData()
cont1='2019/7/8 11:59'
a=ibdata[ibdata.index>cont1]['bias_HSI'].agg(['sum','mean','count'])
b=jqdata[(jqdata.volume>0) & (jqdata.index>cont1)]['bias_IF'].agg(['sum','mean','count'])
pd.concat([a,b],axis=1)
```
