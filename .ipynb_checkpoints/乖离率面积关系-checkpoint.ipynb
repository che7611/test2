```python
import pandas as pd
from KRData.HKData import HKFuture
import numpy as np
import talib
from jqdatasdk import *
```

```python
auth('13580726532', '127565568yjd')
```

```python
get_security_info('IF9999.CCFX')
```

```python
data_if=get_price('IF9999.CCFX', start_date='2019-01-01', end_date='2019-06-19', frequency='1m', fields=None, skip_paused=False, fq=None)
data_if['ma60'] = talib.MA(data_if['close'].values, timeperiod=60)
data_if['bias'] = (data_if['close'].values - data_if['ma60'].values)/data_if['ma60'].values
```

```python
data_if
```

```python
data_if.index = data_if.index - pd.Timedelta(1, 'm')
```

```python
def calc(df):
    try:
        _df = df.between_time('9:30', '11:29')
        return _df['bias'].sum()
    except:
        return np.nan
ins_if = data_if.groupby(data_if.index.date).apply(calc)
```

```python
hf = HKFuture()
```

```python
data_hsi = hf.get_main_contract_bars('HSI', ['open', 'high', 'low', 'close'], start='20190101', end='20190619')
data_hsi['ma60'] = talib.MA(data_hsi['close'].values, timeperiod=60)
data_hsi['bias'] = (data_hsi['close'].values - data_hsi['ma60'].values)/data_hsi['ma60'].values
```

```python
data_hsi
```

```python
ins_hsi = data_hsi.groupby(data_hsi.index.date).apply(calc)
```

```python
ins_m =pd.concat([ins_if, ins_hsi], join='inner',axis=1, sort=True)
```

```python
ins_m.columns = ['HSI', 'IF']
```

```python
%pylab inline
```

```python
ins_m = ins_m[~(ins_m['HSI'].isna()|ins_m['IF'].isna())]
```

```python
ins_m.plot(x='HSI', y='IF', kind='scatter')
```

```python
import statsmodels.api as sm
from statsmodels import regression
```

```python
y = ins_m['HSI'].values
x = ins_m['IF'].values
x = sm.add_constant(x)
```

```python
model = regression.linear_model.OLS(y, x).fit()
```

```python
slope = model.params[0]
beta = model.params[1]
```

```python
_x = np.linspace(ins_m['IF'].values.min(), ins_m['IF'].values.max(), 100)
_y = _x * beta + slope
```

```python
beta, slope
```

```python
plot(ins_m['IF'] * beta - ins_m['HSI'])
```

```python
ins_m.plot(x='IF', y='HSI', kind='scatter')
plot(_x, _y, 'r')
plot(_x, _y + 0.15, 'y')
plot(_x, _y - 0.15, 'g')
```

```python
print(f'IF涨幅对于HSI涨幅的解释度->RSquare:{model.rsquared:.2%}')
```

```python
plt.hist(model.resid, bins=30) # 残差分布
```

$HSI_{\mbox{bias}} = {beta} * IF_{\mbox{bias}} + {slope}$
